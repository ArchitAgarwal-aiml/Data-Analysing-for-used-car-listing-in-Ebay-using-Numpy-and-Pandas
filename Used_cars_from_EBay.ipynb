## Dealing with invalid registration data

Let's determine what percentage of our data has invalid values in this column:

```python
(~auto["registration_year"].between(1900,2016)).sum() / auto.shape[0]
```

Comes out to be less than 4%, we will remove this data

```python
auto = auto[auto["registration_year"].between(1900,2016)]
auto["registration_year"].value_counts(normalize=True).head(10)
```

## Exploring Price by Brand

```python
auto["brand"].value_counts(normalize=True)
```

It seems that volkswagen is most popular brand with approximately double the cars for sale of the next two brands combined..German manufacturers represent four out of the top five brands in most listed cars.

There are lots of brands that have total less than 5% listings ,so we will limit our analysis to brands representing more than 5% of total listings.

```python
brand_counts = auto["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > .05].index
print(common_brands)
```

```python
brand_mean_prices = {}

for brand in common_brands:
    brand_only = auto[auto["brand"] == brand]
    mean_price = brand_only["price"].mean()
    brand_mean_prices[brand] = int(mean_price)

brand_mean_prices
```

### We can see analysis from the data that
* audi, mercedez_benz and bmw are being the most expensive car 
* opel and ford is lowest
* Volkswagen is between two extremeties this may explain its popularity

## Exploring Milage


```python
brand_mean_mileage = {}

for brand in common_brands:
    brand_only = auto[auto["brand"] == brand]
    mean_mileage = brand_only["odometer_km"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)

mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
mean_prices = pd.Series(brand_mean_prices).sort_values(ascending=False)
```

```python
brand_info = pd.DataFrame(mean_mileage,columns=['mean_mileage'])
brand_info
```

```python
brand_info["mean_price"] = mean_prices
brand_info
```

The range of the car does not vary as much like the price of the car. But it seems to be that expensive cars has likely to be have more mileage than less expensive cars.
